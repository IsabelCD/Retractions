```python
import pandas as pd
from tqdm import tqdm
import os
```

```python
import os

folder_path = os.path.join("../thesis_data/filtered_journals/journals0-99")
file_names = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
file_names
```

```python
journals = pd.DataFrame()
```

```python
from pybtex.database.input import bibtex

for file in file_names:
    #open a bibtex file
    print(file)
    parser = bibtex.Parser()
    bibdata = parser.parse_file("../thesis_data/filtered_journals/journals0-99/"+file)

    #list to save dictionaries of entries
    lista_entries = []
    for bib_id in bibdata.entries:
        fields = list(bibdata.entries[bib_id].fields)
        entry = bibdata.entries[bib_id].fields
        
        # create unique dict for each entry
        unique_entry_dict = {}
        for field in fields:
            unique_entry_dict[field] = entry[field]
        
        # apend entries in list
        lista_entries.append(unique_entry_dict)
        
    journals = pd.concat([journals, pd.DataFrame(lista_entries)])
```
